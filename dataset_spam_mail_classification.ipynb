```python
df1['Category'] = df1.Category.map({'ham':0, 'spam':1})
df1
```

```python
df2.drop(['Unnamed: 0', 'label_num'], axis=1, inplace=True)
df2.rename(columns={'label': 'Category', 'text': 'Message'}, inplace=True)
df2['Category'] = df2.Category.map({'ham':0, 'spam':1})
df2
```

```python
df3.drop(['FILE_NAME'], axis=1, inplace=True)
df3.rename(columns={'CATEGORY': 'Category', 'MESSAGE': 'Message'}, inplace=True)
df3
```

```python
df1_spam = df1[df1['Category'] == 1]
df2_spam = df2[df2['Category'] == 1]
df3_spam = df3[df3['Category'] == 1]

df1_ham = df1[df1['Category'] == 0]
df2_ham = df2[df2['Category'] == 0]
df3_ham = df3[df3['Category'] == 0]
```

```python
df1_spam.count(), df2_spam.count(), df3_spam.count()
```

```python
df1_ham = df1_ham.iloc[0:747, :]
df2_ham = df2_ham.iloc[0:1499, :]
df3_ham = df3_ham.iloc[0:1896, :]
```

```python
df_spam = pd.concat([df1_spam, df2_spam, df3_spam], axis=0)
df_spam
```

```python
df_ham = pd.concat([df1_ham, df2_ham, df3_ham], axis=0)
df_ham
```

```python
mail = pd.concat([df_spam, df_ham], axis=0)
mail
```

```python
# from sklearn.utils import shuffle
# mail = shuffle(mail, random_state=None)
mail.reset_index(inplace=True, drop=True)
mail
```

```python
mail.to_csv('mail.csv', index=False)
```
